```python
import hvplot.pandas
import pandas as pd
import folium
from IPython.display import display
```

```python
ev_charging_data = pd.read_csv('cleaned_ev_stations.csv')
```

```python
ev_charging_data
```

```python
charging_map = folium.Map(location=[39.8,-98.6],tiles='OpenStreetMap', zoom_start=4)

for i in range(0,len(ev_charging_data)):
       folium.Marker(
          location=[ev_charging_data.iloc[i]['Latitude'], ev_charging_data.iloc[i]['Longitude']],
          popup=ev_charging_data.iloc[i]['Station Name']).add_to(charging_map)

# for i, row in ev_charging_data.iterrows():
#     Marker(location=[row['Latitude'], row['Longitude']],
#            popup=row['Street Address']).add_to(charging_map)


charging_map.save('charging_map.html') 
display(charging_map)
```
