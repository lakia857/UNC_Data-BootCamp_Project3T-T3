```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
import numpy as np
```

```python
sales_compared_to_load = Path("Resources/car_sales_compared.csv")
```

```python
all_sales = pd.read_csv(sales_compared_to_load)
```

```python
all_sales_final = all_sales[["Entity","Year","Electric cars sold","Non-electric car sales"]]
```

```python
all_sales_final.set_index(['Year'])
```

```python
all_sales_final.groupby(['Entity'])['Electric cars sold'].plot(figsize=(12, 4),legend=True)
pd.set_option('display.float_format', lambda x: '%.0f' % x)

```

```python
gas_prices_to_load = Path("Resources/gas_prices.csv")
gas_prices = pd.read_csv(gas_prices_to_load)
gas_prices.set_index('Year')
```
